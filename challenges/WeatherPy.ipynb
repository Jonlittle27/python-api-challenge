# Observations

1. The first and foremost notable trend is the relationship between the equator and the temperature.  As expected the farther a value travels away from the equator the lower the temperature goes and the closer a value gets to the equator the higher the temperature will be.  This is supported by the northern hemisphere v. temperature graph showing a downward trend in temperature while latitude increased and the southern hemisphere v. temperature graph showed an upward trend in temperature while latitude increased.

2. The next observation is that latitude and wind speed do not appear to be correlated based on the sample gathered.  As seen in the city latitude v. windspeed chart a vast majority of the points fall below the ten miles per hour mark and this is consistent at all latitude points.  Even at the lowest and highest latitudes, places are known to have harsh weather conditions, the points generally fall under ten miles per hour.  The only exception is a few outliers at the highest northern latitudes, but that can just be the result of a storm happening exactly when the data was requested.  

3. Finally, the last observation noted is that the farther north a value traveled the more humid it would become.  Temperature saw a rise toward the equator from the southern hemisphere and then a fall when traveling away from it.  Whereas, humidity starting at the lowest latitude in the southern hemisphere v. humidity plot rose in percentage toward the equator.  Similarly, in the northern hemisphere v humidity chart, the percentage values continued to rise away from the equator and also showing that the highest northern latitude had higher humidity percentages than the lowest southern latitude.   

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress
from pprint import pprint
from datetime import datetime

# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

```python
#Testing the request / to find key and value names 
response = requests.get(f"http://api.openweathermap.org/data/2.5/weather?q=nutley&appid={weather_api_key}").json()
pprint(response)
```

```python
#API Calls Data Retrieval 
print("Beginning Data Retrieval")
print("------------------------")

weather = []

#set empty lists to catch the information (append it all as a dictionary) needed then build a dataframe below

for city in cities:
    try:
        response = requests.get(f"http://api.openweathermap.org/data/2.5/weather?q={city}&appid={weather_api_key}").json()
        weather.append({"city": response["name"],
                        "cloudiness": response["clouds"]["all"],
                        "country": response["sys"]["country"],
                        "date": response["dt"],
                        "humidity": response["main"]["humidity"],
                        "lat": response["coord"]["lat"],
                        "lon": response["coord"]["lon"],
                        "temp_k": response["main"]["temp"],
                        "wind_speed": response["wind"]["speed"]
                       })
                        
        print(f"Processing data for {city}...")
        
        
    except KeyError:
        print("City not found. Skipping...")
    
#Setting a one second delay on requests to avoid the API limit    
    time.sleep(1)
        
print("------------------------")   
print("End of Data Retrieval")                             
                            
        
        
```

```python
weather_data = pd.DataFrame(weather)
weather_data.count()



```

```python
#hemisphere data frame creations
north = []
south = []

for city in weather:
    if city["lat"] > 0:
        north.append(city)
    else:
        south.append(city)
        
northern_hemi = pd.DataFrame(north)
southern_hemi = pd.DataFrame(south)


```

```python
#nothern hemi - lat v temp
x_values = northern_hemi["lat"]
y_values = (northern_hemi["temp_k"]*(9/5))-459.67
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(6,10),fontsize=15,color="red")
plt.title("Northern Hemisphere - Latitude v. Temperature")
plt.xlabel('Latitude')
plt.ylabel('Temperature (f)')
print(f"The r-squared is: {rvalue}")
plt.show()
plt.savefig("Images/north_lat_temp.png")
```
