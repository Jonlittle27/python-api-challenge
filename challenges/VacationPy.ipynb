```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os

# Import API key
from api_keys import g_key

# Configure gmaps
gmaps.configure(api_key=g_key)
```

### Store Part I results into DataFrame
* Load the csv exported in Part I to a DataFrame

```python
weather_data = pd.read_csv("weather_data.csv")
weather_data.head()
```

### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
# Store latitude and longitude in locations/ Humidity for weights
locations = weather_data[["lat", "lon"]].astype(float)
weights = weather_data["humidity"].astype(float)

# Plot Heatmap
fig = gmaps.figure()

# Create heat layer
heat_layer = gmaps.heatmap_layer(locations, weights=weights, 
                                 dissipating=False, max_intensity=weights.max(),
                                 point_radius=1)


# Add layer
fig.add_layer(heat_layer)

# Display figure
fig


```

### Create new DataFrame fitting weather criteria
* Narrow down the cities to fit weather conditions.
* Drop any rows will null values.

```python
#adding a temperature in fahrenheit to the dataframe
weather_data["temp_f"] = (weather_data["temp_k"]*(9/5))-459.67

ideal_weather = weather_data[(weather_data.temp_f > 70) & (weather_data.temp_f <80) 
                             & (weather_data.wind_speed < 10) & (weather_data.cloudiness == 0)]
                                 
ideal_weather
```

### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.
* Plot markers on top of the heatmap.

```python
hotel_df = ideal_weather
```

```python
hotel_df["Hotel Name"] = ""
hotel_df["Hotel Address"] = ""
hotel_df
```

```python
#parameters
params = {"radius": 5000,
    "types": "lodging",
    "key": g_key}

#Looping through each row to find lat/lon 
for index, row in hotel_df.iterrows():
    lat = row["lat"]
    lon = row["lon"]

    #taking the lat/lon and creating the location paramter
    params["location"] = f"{lat},{lon}"

    #URl/response formatting
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
    response = requests.get(base_url, params=params).json()
    
    
    
    try:
        hotel_df.loc[index, "Hotel Name"] = response["results"][0]["name"]
        hotel_df.loc[index, "Hotel Address"] = response["results"][0]["vicinity"]
        

    except (KeyError, IndexError):
        hotel_df.loc[index, "Hotel Name"] = "Unavailable"
        hotel_df.loc[index, "Hotel Address"] = "Unavailable"

#I know the instructions want hotels, but I wasn't sure if that meant exactly hotels or what technically counted so I 
#kept the parameters at just lodging because using keyword "hotel" gave far fewer results
hotel_df
```

```python
# Using the template add the hotel marks to the heatmap
info_box_template = """
<dl>
<dt>Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{city}</dd>
<dt>Country</dt><dd>{country}</dd>
</dl>
"""
# Store the DataFrame Row
hotel_info = [info_box_template.format(**row) for index, row in hotel_df.iterrows()]
locations = hotel_df[["lat", "lon"]]

#heat map 
fig = gmaps.figure()

fig.add_layer(heat_layer)

#Assign the marker layer to a variable
markers = gmaps.marker_layer(locations)

# Add the layer to the map
fig.add_layer(markers)

# Display Map
fig
```
